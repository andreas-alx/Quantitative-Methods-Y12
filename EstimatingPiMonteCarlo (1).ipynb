```python
import random 
import matplotlib.pyplot as plt  # Added import for matplotlib.pyploy

total_points = 5000 # Increase to higher values for greater accuracy
inside_circle = 0

x_inside, y_inside = [], []
x_outside, y_outside = [], []

for _ in range(total_points):
    x, y = random.random(), random.random()

    if x**2 + y**2 <= 1:
        inside_circle += 1 
        x_inside.append(x)
        y_inside.append(y)
    else:
        x_outside.append(x)
        y_outside.append(y)

pi_estimate = (inside_circle / total_points) * 4
print(f"Total Points: {total_points}")
print(f"Points Inside Circle: {inside_circle}")
print(f"My estimate for Pi is: {pi_estimate}")

plt.figure(figsize=(7,7))
plt.scatter(x_inside, y_inside, color='blue', s=1, label="Inside")
plt.scatter(x_outside, y_outside, color='red', s=1, label="Outside")
plt.title(f"Monte Carlo Estimation of Pi: {pi_estimate}")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()


```

Increasing total points causes circle to appear smoother, Pi estimate is closer to 3.141..., Through 50,000 trials an accuracy of 99.9% is achieved.
