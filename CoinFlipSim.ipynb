```python
import random

flips = 1000000
heads = 0

for i in range(flips):
    if random.random() < 0.5:
        heads += 1

percentage = (heads/flips) * 100
print(f"Out of {flips} flips, we got {heads} heads.")
print (f"That is {percentage}% - very close to the theoretical 50%!")
```

I observed as number of trials increase the variance in results decreased, trending towards 50%, which alligns with the Law of Large Numbers. Converging towards the expected value of 0.5.

```python
import matplotlib.pyplot as plt 

labels = ['Heads', 'Tails']
counts = [heads, flips - heads]

plt.bar(labels, counts, color=['blue', 'orange'])
plt.title(f'Results of {flips} Coin Flips')
plt.ylabel('Frequency')
plt.show() 
```
